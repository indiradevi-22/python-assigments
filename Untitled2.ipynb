# Question 1:report **card**

```
maths=int(input("enter maths grade"))
physics=int(input("enter ur ps grade"))
chemistry=int(input("enter yur chemistry grade"))
total=maths+physics+chemistry
grade=(total/3)
if((9<=grade)and(grade<=10)):
  print("your grade is A+")
elif ((7<=grade)and(grade<=8)):
  print("your final grade is b+")
elif ((5<=grade)and(grade<=6)):
  print("your final grade is c+")
else:
  print("you got failed")

```
